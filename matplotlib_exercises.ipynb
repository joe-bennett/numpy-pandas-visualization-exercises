```python
x= list(range(50))
y=np.sqrt(x)
plt.plot(x,y, c= 'red')
plt.title('chart for $y = sqrt(x)$')
plt.show()



x= np.array(range(-50,50))
y=np.power(x,3)
plt.plot(x,y)
plt.title('chart for $y = x^3$')
plt.show()

x= np.array(range(0,100))
y=np.power(2,x)
plt.plot(x,y)
plt.title('chart for $y = 2^x$')
plt.show()



x=np.array(range(0,100))
y=1/(x+1)
plt.plot(x,y)
plt.title('chart for y=1/(x+1)')
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x= list(range(10))
plt.subplot(221)
y=np.sqrt(x)
plt.plot(x,y, c= 'red')
plt.title('chart for $y = sqrt(x)$')




x= np.array(range(0,10))
#y=np.power(x,3)
plt.subplot(222)
y = x**3
plt.plot(x,y)
plt.title('chart for $y = x^3$')


x= np.array(range(0,10))
#y=np.power(2,x)
plt.subplot(223)
y=2**x
plt.plot(x,y)
plt.title('chart for $y = 2^x$')




x=np.array(range(0,10))
plt.subplot(224)
y=1/(x+1)
plt.plot(x,y)
plt.title('chart for y=1/(x+1)')

plt.suptitle ('subplots demo')

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

plt.show()
```
